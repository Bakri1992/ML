##### Train-Test-Split:
Lets create this function our own
  

```python
def train_test_split(df,test_size):
    # We have to check if the test_size we passsed an integer or a float.
    if isinstance(test_size,float):
        test_size=round(test_size*len(df))
    else:
        test_size=test_size
    # df.index returns a range so we can convert it to a list
    indices=df.index.tolist()
    # random.sample accepts in population a list and returns unique values.
    test_indices=random.sample(population=indices,k=test_size)
    test_df=df.loc[test_indices]
    train_df=df.drop(test_indices)
    return train_df,test_df
```

```python
# To get the same result always
random.seed(0)
train_df,test_df=train_test_split(df,test_size=0.1)
```
